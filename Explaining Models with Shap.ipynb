```python
## Our standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
## Preprocessing tools
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_transformer, make_column_selector
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
## Models & evaluation metrics
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
## setting random state for reproducibility
SEED = 321
np.random.seed(SEED)
```

```python
def evaluate_regression(model, X_train,y_train, X_test, y_test): 
    """Evaluates a scikit learn regression model using r-squared and RMSE"""
    
    ## Training Data
    y_pred_train = model.predict(X_train)
    r2_train = metrics.r2_score(y_train, y_pred_train)
    rmse_train = metrics.mean_squared_error(y_train, y_pred_train, 
                                            squared=False)
    
    print(f"Training Data:\tR^2= {r2_train:.2f}\tRMSE= {rmse_train:.2f}")
        
    
    ## Test Data
    y_pred_test = model.predict(X_test)
    r2_test = metrics.r2_score(y_test, y_pred_test)
    rmse_test = metrics.mean_squared_error(y_test, y_pred_test, 
                                            squared=False)
    
    print(f"Test Data:\tR^2= {r2_test:.2f}\tRMSE= {rmse_test:.2f}")
```

## Load Your Saved Joblib File

```python
import joblib
loaded = joblib.load('random_forest_l01.joblib')
loaded.keys()
```

```python
X_train_df = loaded['X_train']
y_train = loaded['y_train']
X_test_df = loaded['X_test']
y_test = loaded['y_test']
preprocessor = loaded['preprocessor']
loaded_model = loaded['RandomForest']
```

```python
evaluate_regression(loaded_model,X_train_df, y_train, X_test_df,y_test)
```

## Explain your tree-based model with shap

```python
import shap
shap.initjs()
```

```python
X_shap = shap.sample(X_train_df, nsamples=200, random_state=321)
X_shap
```

```python
## get the corresponding y-values
y_shap = y_train.loc[X_shap.index]
y_shap
```

```python
explainer = shap.Explainer(loaded_model)
explainer
```

```python
shap_values = explainer(X_shap,y_shap)
shap_values.shape
```

```python
import os
os.makedirs("images/", exist_ok=True)
```

```python
fig, ax = plt.subplots()
shap.summary_plot(shap_values,X_shap, plot_type='bar')
fig.savefig('images/summary_plot_bar.png')
```
